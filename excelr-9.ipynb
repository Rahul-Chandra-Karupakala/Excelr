Assignment-2:  Data Aggregation and Grouping

```python
grouped_by_account_type = df.groupby('Account_Type').agg(
    Total_Transaction_Amount=('Transaction_Amount', 'sum'),
    Average_Account_Balance=('Account_Balance', 'mean')
).reset_index()
grouped_by_account_type
```

```python
grouped_by_branch = df.groupby('Branch').agg(
    Total_Transactions=('Transaction_Type', 'count'),
    Total_Transaction_Amount=('Transaction_Amount', 'sum'),
    Average_Transaction_Amount=('Transaction_Amount', 'mean'),
    Average_Account_Balance=('Account_Balance', 'mean')
).reset_index()
grouped_by_branch
```

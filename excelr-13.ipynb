Assignment 1: Sales and Effectiveness Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("Pharma_data.csv")
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
region_sales = df.groupby("Region")["Sales"].sum()
plt.figure(figsize=(8, 5))
region_sales.plot(kind="bar", color="skyblue")
plt.title("Total Sales per Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Marketing_Spend", y="Sales", hue="Region")
plt.title("Marketing Spend vs Sales")
plt.xlabel("Marketing Spend")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Age_Group", y="Effectiveness", palette="Set2")
plt.title("Drug Effectiveness Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Effectiveness")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="Trial_Period", y="Sales", hue="Product_Name", marker="o", palette="Set1")
plt.title("Sales Trend for Each Product Over Trial Periods")
plt.xlabel("Trial Period")
plt.ylabel("Sales")
plt.legend(title="Product Name", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df[["Sales", "Marketing_Spend", "Effectiveness"]].corr()
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Sales, Marketing Spend, and Effectiveness")
plt.show()
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("Pharma_data.csv")
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x="Region", y="Effectiveness", hue="Product_Name", ci=None, palette="Set2")
plt.title("Average Effectiveness for Each Drug Across Different Regions")
plt.xlabel("Region")
plt.ylabel("Effectiveness")
plt.legend(title="Product Name", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x="Product_Name", y="Effectiveness", inner="quartile", color="skyblue", bw=0.2)
sns.violinplot(data=df, x="Product_Name", y="Side_Effects", inner="quartile", color="lightcoral", bw=0.2, alpha=0.6)
plt.title("Distribution of Effectiveness and Side Effects for Each Product")
plt.xlabel("Product Name")
plt.ylabel("Values")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df, vars=["Effectiveness", "Side_Effects", "Marketing_Spend"], hue="Product_Name", palette="Set1")
plt.suptitle("Pairplot of Effectiveness, Side Effects, and Marketing Spend", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Trial_Period", y="Effectiveness", palette="pastel")
plt.title("Effectiveness Across Different Trial Periods")
plt.xlabel("Trial Period")
plt.ylabel("Effectiveness")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x="Marketing_Spend", y="Effectiveness", scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title("Effect of Marketing Spend on Drug Effectiveness")
plt.xlabel("Marketing Spend")
plt.ylabel("Effectiveness")
plt.tight_layout()
plt.show()
```
